```python
import json
import re
import pickle
import pandas as pd
import numpy as np
```

```python
with open('LOCATION_TO_COUNTRY_MAP.pickle', 'rb') as handle:
    mapping = pickle.load(handle)
```

```python
mapping
```

```python
with open('FAILED.pickle', 'rb') as handle:
    failed = pickle.load(handle)
```

```python
failed
```

```python
all_dicts = []

for string_data in failed:
    formatted_data = string_data.encode('utf-8').decode('unicode_escape')
    
    try:
        dict_list = json.loads(formatted_data)
        all_dicts.append(dict_list)
    except json.JSONDecodeError as e:
        print(f"Failed to parse JSON: {e}")
```

```python
all_dicts
```

```python
all_dicts = [d for sublist in all_dicts for d in sublist]
all_dicts
```

```python
addl_dict = {}
for loc_dict in all_dicts: 
    addl_dict[loc_dict["location"]] = loc_dict["country"]
addl_dict
```

```python
len(mapping), len(addl_dict)
```

```python
mapping = mapping | addl_dict
```
