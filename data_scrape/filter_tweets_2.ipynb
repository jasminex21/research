```python
ft_indicators = ["FULL TIME", "FULL-TIME", "Full time", "Full-time"]
```

```python
# get the tweet ids of tweets AFTER the full time tweets
parent_tweets[parent_tweets.tweet.str.contains("FULL-TIME")]
```

```python
full_time_pattern = "|".join(ft_indicators)
full_time_tweets = parent_tweets[parent_tweets['tweet'].str.contains(full_time_pattern, case=False)]

later_tweet_ids = []

for _, full_time_row in full_time_tweets.iterrows():
    full_time_date = full_time_row['datetime'].date()  # Get the date of the "FULL-TIME" tweet
    full_time_timestamp = full_time_row['datetime']  # Get the exact timestamp of the "FULL-TIME" tweet
    full_time_club = full_time_row["club_handle"]

    # Filter tweets on the same day with a later timestamp
    same_day_later_tweets = parent_tweets[
        (parent_tweets["club_handle"] == full_time_club) &
        (parent_tweets['datetime'].dt.date == full_time_date) & 
        (parent_tweets['datetime'] > full_time_timestamp) 
    ]
    
    later_tweet_ids.extend(same_day_later_tweets['tweet_id'].tolist())
```

```python
# give each parent tweet a kickoff time
parent_tweets["ko_time"] = [club_stats[club_stats["report_id"] == rep_id]["date"].values[0] + " " + club_stats[club_stats["report_id"] == rep_id]["time"].values[0] for rep_id in parent_tweets["report_id"]]
```

```python
parent_tweets["mins_after_ko"] = ((parent_tweets['datetime'] - parent_tweets['ko_time']).dt.total_seconds() / 60).astype(int)
```

```python
def convert_minute(minute):
    if '+' in minute:
        base_minute, added_time = minute.split('+')
        return int(base_minute) + int(added_time)
    else:
        return int(minute)
```

```python
events["minute_eng"] = events['minute'].apply(convert_minute)
```

```python
events
```

```python
events[events["game_id"] == "2ba17e6d"]
```

```python
def add_ht(minute):
    if int(minute) > 45:
        return int(minute) + 15
    else:
        return int(minute)
```

```python
events["minute_eng"] = events['minute_eng'].apply(add_ht)
```
